#Merge & Clean Data

```python
#Make year Dimension
year = pd.read_excel('/content/drive/MyDrive/yeardim.xlsx')
hj = []
for i in year['Year'].astype('str'):
  hj.append(i)
hj.append(" ")
#load data
artists = "https://media.githubusercontent.com/media/MuseumofModernArt/collection/master/Artists.csv"
artwork = "https://media.githubusercontent.com/media/MuseumofModernArt/collection/master/Artworks.csv"

artists = pd.read_csv(artists)
artwork = pd.read_csv(artwork)
#merge data
df= artists.merge(artwork, left_on = "DisplayName", right_on= "Artist")
#define cleaning functions
def remove_characters(col):
    return re.sub(r"[a-zA-Z]",'',col)

parent1 = ")"
parent2 = "("
def remove_text(col):
  if parent1 or parent2 in col:
    return re.sub(r"[.,/)()]",'',col)
def remove_scores(col):
    return re.sub(r"-",'',col)



df['Date'] = df['Date'].astype('str')
#Clean Dates
df['Date_New'] = df['Date'].apply(lambda x:remove_characters(x))
df['Date_New'] = df['Date_New'].apply(lambda x:remove_text(x))
df['Date_New'] = df['Date_New'].apply(lambda x:remove_scores(x))
df['Date_New'] = df['Date_New'].str[-4:].fillna(0)
#df.merge(year['Year'],how='inner',left_on ='Date_New', right_on="Year")
final = pd.concat([df, year], axis=1, sort=False)


#Only use non messy dates
#Clean up Messy Dates (Remove around 14k records for sanitation)
df = df[df['Date_New'].isin(hj)]
final.info()
df.info()
```

##Top Artist's Work Over Time by Type

```python
#Get Rid of Unkown Artists
art = df[~df['Artist'].str.contains('Unknown|Various')]

#group by Artist, Data and count of new Titles created (take top 100)
go = art.groupby(['Artist'])['Title'].count().sort_values(ascending=False).head(100)
#reset index
go = go.to_frame().reset_index()
plt.figure(figsize=(20,5))
#get list of top artits
top_artist = []
for i in go['Artist'].unique():
  top_artist.append(i)

#create a data frame dictionary to store your data frames
DataFrameDict = {elem : pd.DataFrame for elem in top_artist}
#create seperate data frame for each corresponding key(artist name)
for key in DataFrameDict.keys():
    DataFrameDict[key] = df[:][df.Artist == key]
#create function that plots data using plotly based on Date, Classification, and Title Count

```

```python
def groupby(df):
  g = df.groupby(['Date_New','Classification'])['Title'].count().sort_values(ascending=False)
  g = g.reset_index()
  fig = px.bar(g.sort_values(by="Date_New"), x= "Date_New", y = 'Title', color = "Classification",title = "{}'s Work Over Time By Type".format((df['DisplayName'].unique())))
  fig.update_layout(template='ggplot2')
  fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.01
))
  return fig.show()
#use list comprehension to pipe groupby function to data frame dictionary 
df_dict = {k: v.pipe(groupby) for k, v in DataFrameDict.items()}
```
